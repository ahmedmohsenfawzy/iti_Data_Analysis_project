```python
#Show all students who Passed and have Grade ≥ 85.
df[(df['Passed'] == True) & (df['Grade'] >= 85)]
```

```python
#Show students whose Name contains the letter "a" or "A" (case insensitive)
df[df['Name'].str.contains('a', case=False)]
```

```python
#Show students in Cairo or Alexandria with Grade < 80.
df[(df['City'].isin(['Cairo', 'Alexandria'])) | (df['Grade'] < 80)]
```

```python
#Show students with Grade between 80 and 90 (inclusive).
df[df['Grade'].between(80,90)]
```

```python
#Using .loc, change Khaled's grade from 55 to 68.
df.loc[df['Name'] == 'Khaled', ['Grade']] = 68
```

```python
df['Grade']
```

```python
#Add a new column Is_Excellent that is True if Grade ≥ 90, False otherwise (vectorized).
df['Is_Excellent'] = df['Grade'].map(lambda x:True if x >= 90 else False)
```

```python
df
```

```python
#Fill missing Grade values with the average Grade
df['Grade'].fillna(value=df['Grade'].mean(), inplace=True)
```

```python
df['Score Date'].fillna(value=df['Score Date'].mode()[0], inplace=True)
```

```python
#•Add a new column Final_Grade = Grade + 5 (bonus points).
df['Final_Grade'] = df['Grade'] + 5
```

```python
#Create a column Grade_Category using pd.cut with bins:
# •0–69 → "Fail"
# •70–79 → "Average"
# •80–89 → "Good"
# •90–100 → "Excellent"
bin = [0,70,80,90,100]
label = ["Fail", "Average", "Good", "Excellent"]
df['Grade_Category'] = pd.cut(df['Grade'], bins=bin , labels=label)
```

```python
df
```

```python
# •Remove the column Student ID.
df.drop(columns=['Student ID'], inplace=True)
```

```python
# •Keep only students who Passed.
df = df[df['Passed'] == True]
```

```python
# •Sort the DataFrame by Grade descending.
df.sort_values(by='Grade', ascending=False)
```

```python
#•Sort by City ascending, then inside each city sort by Grade descending.
df.sort_values(by=['City','Grade'], ascending=[True,False])
```

```python
#•Add a column Rank that ranks students by Grade (1 = highest, handle ties with 'min').
df['Rank'] = df['Grade'].rank(ascending=True, method='dense')
```

```python
df
```

```python
#•Show the top 4 students with the highest grades (use nlargest).
df.nlargest(4,'Rank',keep='first')
```

```python
# •Find the student(s) with the highest Grade and show their Name and Grade.
df[df['Grade'] == df['Grade'].max()][['Name', 'Grade']]
```

```python
#•After filling missing grades, recalculate the average Grade per Subject.
df.groupby('Subject')['Grade'].mean()
```

```python
#•Create a new column City_Avg that shows the average grade of each student's city (using transform).
df['City_Avg'] = df.groupby('City')['Grade'].transform('mean')
```

```python
#•Show the subject with the highest average grade.
df[df['Grade'] == df['Grade'].max()]['Subject']
```
